```python
import wikipedia
import numpy as np
import pandas as pd
import math
```

```python
data = pd.read_csv('../data/FinalSource_Real Cases.csv', sep=',',  error_bad_lines=False, encoding = 'ISO-8859-1')
```

```python
def extract_company(url):
    try : 
        try : 
            company = url.split('www.')[1]
        except : 
            company = url.split('//')[1]
        
        company = company.split('.')[0]
        return company
    except : 
        return np.nan
```

```python
data['company'] = data['URL'].apply(extract_company)
```

```python
def lower(x) : 
    try : 
        x = x.lower()
        return x
    except : 
        return x
    
data['Country'].apply(lambda x : lower(x)).unique()
```

```python
import json
import os
import glob
cwd = os.getcwd()
list_file = glob.glob("{}/norse_json/*.json".format(cwd))
```

```python
def load_data(list_file):
    with open(list_file[0]) as json_data :
        data = pd.DataFrame.from_dict(json.load(json_data))

    for file in list_file[1:] : 
        with open(file) as json_data:
            data = pd.concat([data, pd.DataFrame.from_dict(json.load(json_data))])
    return data
```
