```python
import numpy as np
import pandas as pd

import plotly.plotly as py
import plotly.graph_objs as go

from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('../data/FinalSource_Real Cases.csv', sep=',',  error_bad_lines=False, encoding = 'ISO-8859-1')
```

```python
data
```

```python
data = data[['Date','Notify','URL','Country','OS']]
data['Date'] = pd.to_datetime(data['Date'])
data['day'] = data['Date'].dt.day
data['month'] = data['Date'].dt.month
data['year'] = data['Date'].dt.year
```

```python
data = data.drop_duplicates(subset=['Date','Notify','URL'], keep='first')
```

#### Recognize .gov sites :

```python
def recognize_pattern(url, pattern): 
    if pattern in url : 
        return(1)
    else :
        return(0)
    
data['gov'] = data[data['URL'].notnull()]['URL'].apply(lambda x : recognize_pattern(x, '.gov'))
```

```python
def get_data_chart(x,y):
    data_chart = [go.Bar(
                x=x,
                y=y,
                textposition = 'auto',
                marker=dict(
                    color='rgb(158,202,225)',
                    line=dict(
                        color='rgb(8,48,107)',
                        width=1.5),
                ),
                opacity=0.6
            )]
    return(data_chart)
```

#### Attacks on sites per day 

```python
data_group_date = data.groupby(['Date']).count()
data_chart = get_data_chart(data_group_date.index,data_group_date['URL'])
iplot(data_chart, filename='bar-direct-labels')
```

#### Attacks on .gov sites per year

```python
data_group_date = data.groupby(['year']).count()
data_chart = get_data_chart(data_group_date.index,data_group_date['URL'])
iplot(data_chart, filename='bar-direct-labels')
```

```python
data_group_date = data.groupby(['Country']).count()
data_chart = get_data_chart(data_group_date.index,data_group_date['URL'])
iplot(data_chart, filename='bar-direct-labels')
```
